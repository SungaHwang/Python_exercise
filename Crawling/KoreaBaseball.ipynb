```python
from bs4 import BeautifulSoup
from selenium import webdriver
import selenium
import time
```

```python
driver = webdriver.Chrome('/chromedriver.exe')
driver.get('https://www.koreabaseball.com/Record/Player/Defense/Basic.aspx')
```

```python
baseball_list = []
```

```python
try:
    for i in range(5):
        
        xpath = '//*[@id="cphContents_cphContents_cphContents_ucPager_btnNo'+str(i+1)+'"]'
        print(xpath)
        btn = driver.find_element_by_xpath(xpath)
        print(btn)   
        btn.click()
        
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        data = soup.select('#cphContents_cphContents_cphContents_udpContent > div.record_result > table > tbody')
        rows = data[0].findAll('tr')
        
        for row in rows:
            tds = row.findAll('td')
            
            col0 = tds[0].text
            col1 = tds[1].text
            col2 = tds[2].text
            col3 = tds[3].text
            col4 = tds[4].text
            col5 = tds[5].text
            col6 = tds[6].text
            col7 = tds[7].text
            col8 = tds[8].text
            col9 = tds[9].text
            col10 = tds[10].text
            col11 = tds[11].text
            col12 = tds[12].text
            col13 = tds[13].text
            col14 = tds[14].text
            col15 = tds[15].text
            
            baseball_list.append([col0, col1, col2, col3, col4, col5, col6, col7, col8, col9, col10, col11, col12, col13, col14, col15])
            
        time.sleep(1)
        
        
except:
    print('1')
```

```python
print(baseball_list)
```
