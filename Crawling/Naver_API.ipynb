# Api 이용하여 데이터 수집

#### 네이버 api를 이용하여 '코로나' 키워드로 뉴스 100개 수집

```python
import os
import sys
import urllib.request
import requests
from urllib.parse import urlparse
import json
import pandas as pd
```

```python
client_id= ""
client_secret= ""
```

```python
keyword='코로나'
```

```python
url = "https://openapi.naver.com/v1/search/news.json"
query = "?query=" + keyword
option = "&display=100&start=1&sort=sim"
```

```python
url_all = url+ query+ option
url_all
```

```python
result = requests.get(urlparse(url_all).geturl(),
                      headers={"X-Naver-Client-Id": client_id,
                               "X-Naver-Client-Secret": client_secret})
```

```python
print(result)
```

```python
data= result.json()
```

```python
result.json()['items']
```

```python
title=[]
originallink=[]
link=[]
description=[]
pubDate=[]

for i in data['items']:
    title.append(i['title'].replace('<b>','').replace('</b>','').replace('&quot','').replace(';',''))
    originallink.append(i['originallink'])
    link.append(i['link'])
    description.append(i['description'].replace('<b>','').replace('</b>','').replace('&quot','').replace(';',''))
    pubDate.append(i['pubDate'])
```

```python
df=pd.DataFrame([title,originallink,link,description,pubDate]).T
df.columns=['title','originallink','link','description','pubDate']
df
```

```python
df.to_csv(f'navernews_corona.csv')
```
